I used the sample flights dataset to create some graphs to show relationships between different fields. I was interested in examining the relationships between cities and weather, so I created two graphs based on origin cities and destination cities.

The destination cities and weather graph is interesting. Clear weather is associated with the largest number of cities. Shanghai is the only city that is associated with 3 different weather types: Clear, Thunder & Lightning, and Sunny. Barcelona is the only destination city that is not associated with any weather type, which is odd. There may be data quality issues with missing data.

For some reason, rain is the only weather type which has strong associations with origin cities. Chitose, Beijing, Chicago, and Osaka have the strongest associations with rain. The origin graph is not as interesting as the destination graph. There may be data quality issues where either the OriginCityName or OriginWeather is missing which doesn't allow the graph to be created correctly.
